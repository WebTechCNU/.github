```python
import json
import os

# Path to your JSON file
JSON_FILE = "results.json"
OUTPUT_DIR = "html"

# Create folder if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load JSON data
with open(JSON_FILE, "r", encoding="utf-8") as f:
    students = json.load(f)

# HTML template
HTML_TEMPLATE = """<!DOCTYPE html>
<html lang="uk">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{name}</title>
    <link rel="stylesheet" href="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/css/bootstrap.min.css">
    <script src="https://cdn.jsdelivr.net/npm/jquery@3.6.3/dist/jquery.slim.min.js"></script>
    <script src="https://cdn.jsdelivr.net/npm/popper.js@1.16.1/dist/umd/popper.min.js"></script>
    <script src="https://cdn.jsdelivr.net/npm/bootstrap@4.6.2/dist/js/bootstrap.bundle.min.js"></script>
</head>
<body class="bg-light">
    <nav class="navbar navbar-expand-sm bg-light fixed-top">
            <ul class="navbar-nav">
                <li class="nav-item ">
                    <a class="nav-link btn btn-dark" href="https://github.com/WebTechCNU">To the git org</a>
                </li>
                <h4 class="ml-4 mt-1">Результати навчання за перший модуль</h4>
            </ul>
        </nav>
        <br>
    <div class="container py-5">
        <h2 class="text-center mb-4">{name}</h2>
        <table class="table table-hover table-bordered table-striped">
            <thead class="table-primary">
                <tr><th>Дисципліна</th><th>Бали</th><th>max</th><th>Пропуски</th></tr>
            </thead>
            <tbody>
                <tr><td>Актуальні питання історії та культури України</td><td></td> <td>30</td><td></td></tr>
                <tr><td>Алгебра і геометрія</td><td></td><td>20</td><td></td></tr>
                <tr><td>Дискретна математика</td><td></td><td>30</td><td></td></tr>
                <tr><td>Математичний аналіз</td><td></td><td>28</td><td></td></tr>
                <tr><td>Програмування</td><td></td><td>25</td><td></td></tr>
                <tr><td>Українська мова</td><td></td><td>20</td><td></td></tr>
            </tbody>
        </table>
        <div class="mt-4">
            <h5>Контактна інформація</h5>
            <p>mathmod@chnu.edu.ua</p>
        </div>
    </div>
    <script src="../script.js"></script>
    <script> init("{id}"); </script>
</body>
</html>
"""

# Generate HTML files
for student in students:
    html_content = HTML_TEMPLATE.format(name=student["name"], id=student["id"])
    output_path = os.path.join(OUTPUT_DIR, f"{student['id']}.html")

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(html_content)

print(f"✅ Created {len(students)} HTML files in '{OUTPUT_DIR}' folder.")
```

```python
import os
from weasyprint import HTML

html_folder = "html"
pdf_folder = "pdf"

os.makedirs(pdf_folder, exist_ok=True)

for file in os.listdir(html_folder):
    if file.endswith(".html"):
        html_path = os.path.join(html_folder, file)
        pdf_path = os.path.join(pdf_folder, file.replace(".html", ".pdf"))
        print(f"Converting {file} -> {pdf_path}")
        HTML(filename=html_path).write_pdf(pdf_path)

print("✅ All HTML files converted to PDF!")
```
