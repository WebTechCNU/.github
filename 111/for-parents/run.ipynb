# Generate html files

```python
import json
import os

# Path to your JSON file
JSON_FILE = "results.json"
OUTPUT_DIR = "html"

# Create folder if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Load JSON data
with open(JSON_FILE, "r", encoding="utf-8") as f:
    students = json.load(f)

# HTML template
HTML_TEMPLATE = """<!DOCTYPE html>
<html lang="uk">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{name}</title>
    <link href="https://cdn.jsdelivr.net/npm/bootstrap@5.3.2/dist/css/bootstrap.min.css" rel="stylesheet">
</head>
<body class="bg-light">
    <div class="container py-5">
        <h2 class="text-center mb-4">{name}</h2>
        <table class="table table-hover table-bordered table-striped">
            <thead class="table-primary">
                <tr><th>Дисципліна</th><th>Бали</th><th>max</th><th>Пропуски</th></tr>
            </thead>
            <tbody>
                <tr><td>Актуальні питання історії та культури України</td><td></td> <td>30</td><td></td></tr>
                <tr><td>Алгебра і геометрія</td><td></td><td>20</td><td></td></tr>
                <tr><td>Дискретна математика</td><td></td><td>30</td><td></td></tr>
                <tr><td>Математичний аналіз</td><td></td><td>20</td><td></td></tr>
                <tr><td>Програмування</td><td></td><td>25</td><td></td></tr>
                <tr><td>Українська мова</td><td></td><td>20</td><td></td></tr>
            </tbody>
        </table>
        <div class="mt-4">
            <h5>Контактна інформація</h5>
            <p>mathmod@chnu.edu.ua</p>
        </div>
    </div>
    <script src="../script.js"></script>
    <script> init("{id}"); </script>
</body>
</html>
"""

# Generate HTML files
for student in students:
    html_content = HTML_TEMPLATE.format(name=student["name"], id=student["id"])
    output_path = os.path.join(OUTPUT_DIR, f"{student['id']}.html")

    with open(output_path, "w", encoding="utf-8") as f:
        f.write(html_content)

print(f"✅ Created {len(students)} HTML files in '{OUTPUT_DIR}' folder.")
```
